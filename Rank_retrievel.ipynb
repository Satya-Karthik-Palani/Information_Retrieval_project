# Ranked Retrieval system of hindi captions

```python
import pandas as pd

dataframe = pd.read_excel('ConceptualCaptionsDataset.xlsx',header = None)

first_column = dataframe.iloc[:10, 0]
print("First 10 documents in input dataset")
print(first_column) 
print("The shape of the input dataset is")
print(dataframe.shape)
```

# Printing the first 2000 documents in the input dataset

```python
N=2000
original_documents = []
documents = []
for i in range(N):
    original_documents.append(dataframe.iloc[i, 0])
    documents.append(dataframe.iloc[i, 0])
    print(dataframe.iloc[i, 0])

```

# Preprocessing of dataset

```python
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

sW=set(stopwords.words('english'))
# print(sW)
def remove_punctuations(InputText):
    punctuations = list(string.punctuation)
    punctuations.remove('.')
    for i in range(len(punctuations)):
        InputText=InputText.replace(punctuations[i]," ")
    return InputText

for i in range(N):
    documents[i]=remove_punctuations(documents[i])
    ws=documents[i].split()
    sentence = ""
    for word in ws:
        if word in sW:
            continue
        else:
            sentence=sentence+" "+word
    documents[i]=sentence
#     print(documents[i])
```

# INDEXING

# Bag of words 

```python
all_words=[]
words=[]
for i in range(N):
    words=documents[i].split()
    all_words=words+all_words
print(set(all_words))
print(len(set(all_words)))
```

# TF Vector function

```python
def tf_vector(string_inp):
    my_string = string_inp.lower().split()
    dictionary = {}
    for word in set(all_words):
        dictionary[word] = my_string.count(word)
    return dictionary
```

# Printing TF Vector for first document in dataset

```python
print(tf_vector(documents[0]))
```

# Doc Frequencies of all unique words

```python
final_Dictionary={}
for word in set(all_words):
    final_Dictionary[word] = 0
for i in range(N):
    dictionary = tf_vector(documents[i])
    for word in set(all_words):
        if(dictionary[word]!=0):
            final_Dictionary[word]=final_Dictionary[word]+1
#     final_Dictionary[word] = dictionary[word] + final_Dictionary[word]
print(final_Dictionary)
```

# Calculating IDF

```python
import math
idf_dictionary = {}
for word in set(all_words):
    idf_dictionary[word] = 0
for word,n in final_Dictionary.items():
    if n!=0:
        idf_dictionary[word] = math.log(N/n,2)
print(idf_dictionary)
```

```python
tf_dictionary = []
for i in range(N):
    tf_dictionary.append(tf_vector(documents[i]))
```

```python
import copy
weight =  copy.deepcopy(tf_dictionary)
for i in range(len(weight)):
    for word,tf in tf_dictionary[i].items():
        weight[i][word] = tf * idf_dictionary[word]
```

# Printing weights of unique words present in doc1

```python
print(weight[0])
```

# SEARCHING

# Taking query input

```python
print("Enter the query:")
query = input()
```

# Term frequency of query

```python
tf_query = tf_vector(query)
print(tf_query)
```

# Printing weights of query

```python
wf_query = copy.deepcopy(tf_query)
for word,tf in tf_query.items():
    wf_query[word] = tf * idf_dictionary[word]
print(wf_query)
```

# Calcuting Euclidien distances

```python
def compute_eucliden_dist(input_weight):
    dist = 0
    for word,tf in wf_query.items():
        dist = dist + pow(tf - input_weight[word] , 2)
    return math.sqrt(dist)
```

```python
euclidean_dist = []
for i in range(N):
    euclidean_dist.append(compute_eucliden_dist(weight[i]))
print(euclidean_dist)
```

# Computing similarity coefficients b/w query and doucment

```python
similarity = []
for i in range(N):
    similarity.append(1 / (1 + euclidean_dist[i]))
print(similarity)
```

# Top 10 ranked retrieved documents using "SC"

```python
print("The entered query is ","'" + query + "'")
print("\n------------------------Top 10 ranked documents using similarity coefficient------------------------\n")
import numpy 
array_similarity_list = numpy.array(similarity)
sort_index = numpy.argsort(array_similarity_list)
for i in range(len(sort_index)-1,len(sort_index)-11,-1):
    print(dataframe.iloc[sort_index[i],0],"- index - ",sort_index[i])
    print(dataframe.iloc[sort_index[i],1])
    print()
```

# Computing Similarity using cosine similarity

```python
from scipy import spatial

cosine_similarity = list()
for i in range(N):
    cosine_similarity.append(1 - spatial.distance.cosine(list(wf_query.values()), list(weight[i].values())))
print(cosine_similarity)
```

# Top 10 ranked retrieved documents using cosine similarity

```python
print("The entered query is ","'" + query + "'")
print("\n------------------------Top 10 ranked documents using cosine similarity------------------------\n")
import numpy 
array_similarity_list = numpy.array(cosine_similarity)
sort_index = numpy.argsort(cosine_similarity)
for i in range(len(sort_index)-1,len(sort_index)-11,-1):
    print(dataframe.iloc[sort_index[i],0],"-index - ",sort_index[i])
    print(dataframe.iloc[sort_index[i],1])
    print()
```
